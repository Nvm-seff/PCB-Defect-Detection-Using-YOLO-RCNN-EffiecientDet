```python
# ==============================================================================
# 0. Common Setup & Imports
# ==============================================================================
import os
import shutil
import random
import xml.etree.ElementTree as ET
import json
import time
from pathlib import Path
from collections import defaultdict

import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import torch
import torchvision

# <<< Ensure necessary torchvision components are imported >>>
from torchvision.models.detection import FasterRCNN
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
# Use either v1 or v2 transforms consistently. Let's stick with v2 as used before.
from torchvision.transforms.v2 import functional as F, ConvertImageDtype, Compose, RandomHorizontalFlip
# from torchvision.transforms import functional as F # Use this for v1 ToTensor
# from torchvision.transforms import ToTensor, Compose, RandomHorizontalFlip # v1 imports
# <<< END Ensure necessary torchvision components are imported >>>

from torch.utils.data import DataLoader, Dataset
from tqdm.notebook import tqdm  # Use notebook version if in Colab/Jupyter

# --- Ultralytics YOLO ---
# <<< Ensure necessary Ultralytics components are imported >>>
from ultralytics import YOLO
import yaml
# <<< END Ensure necessary Ultralytics components are imported >>>


# --- TensorFlow (for EfficientDet suggestion) ---
# import tensorflow as tf # Keep commented if not used yet
# import tensorflow_hub as hub # Keep commented if not used yet

# --- Plotting & Metrics ---
import seaborn as sns
from sklearn.model_selection import train_test_split
# For mAP calculation (using torchmetrics for PyTorch models)
try:
    import torchmetrics
    from torchmetrics.detection import MeanAveragePrecision
except ImportError:
    print("torchmetrics not found. Install using: pip install torchmetrics pycocotools")
    MeanAveragePrecision = None


# --- Configuration ---
# <<< Use Kaggle input/output directories >>>
ROOT_DIR = Path('/kaggle/input') # Standard Kaggle input
KAGGLE_DATASET_DIR = ROOT_DIR / 'pcb-defects' # Assuming dataset is named 'pcb-defects'

BASE_DATA_DIR = KAGGLE_DATASET_DIR / 'PCB_DATASET' # Intermediate directory observed in path fix
ANNOTATIONS_DIR = BASE_DATA_DIR / 'Annotations'
IMAGES_DIR = BASE_DATA_DIR / 'images'

# Output directory for saved files (models, processed data)
OUTPUT_DIR = Path('/kaggle/working') # Standard Kaggle output, persistent after commit
# <<< END Use Kaggle input/output directories >>>

YOLO_DATA_DIR = OUTPUT_DIR / 'yolo_data'
FASTER_RCNN_DATA_DIR = OUTPUT_DIR / 'faster_rcnn_data'
# EFFICIENTDET_DATA_DIR = OUTPUT_DIR / 'efficientdet_data' # Not saving model for this
MODEL_CHECKPOINTS_DIR = OUTPUT_DIR / 'model_checkpoints' # Directory to store final best models

# Create directories safely
OUTPUT_DIR.mkdir(exist_ok=True)
YOLO_DATA_DIR.mkdir(exist_ok=True)
FASTER_RCNN_DATA_DIR.mkdir(exist_ok=True)
MODEL_CHECKPOINTS_DIR.mkdir(exist_ok=True) # Ensure this directory exists

# --- Constants ---
CLASSES = ['missing_hole', 'mouse_bite', 'open_circuit', 'short', 'spur', 'spurious_copper']
NUM_CLASSES = len(CLASSES)
CLASS_MAP = {name: i for i, name in enumerate(CLASSES)}
INV_CLASS_MAP = {i: name for i, name in enumerate(CLASSES)} # For visualization

SEED = 42
IMG_SIZE_YOLO = 640 # Standard for YOLOv8
IMG_SIZE_RCNN = 640 # Can adjust, but keep consistent if possible
BATCH_SIZE_YOLO = 16 # Adjust based on GPU memory (Kaggle T4/P100 usually handles 16 for yolov8s)
BATCH_SIZE_RCNN = 4  # Faster R-CNN often requires smaller batch size
EPOCHS_YOLO = 50 # Adjust as needed
EPOCHS_RCNN = 25 # Adjust as needed
LEARNING_RATE_RCNN = 0.001
DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
print(f"Using device: {DEVICE}")

# Set seed for reproducibility
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ==============================================================================
# 1. Data Preparation (Common Functions)
# ==============================================================================

# <<< Functions parse_xml, find_image_path, load_all_annotations are unchanged from your fixed version >>>
# ... (Keep the fixed versions of parse_xml, find_image_path, load_all_annotations here) ...
def parse_xml(xml_file, base_img_dir):
    """Parses a single XML file and returns image info and annotations."""
    try:
        tree = ET.parse(xml_file)
        root = tree.getroot()

        filename_node = root.find('filename')
        size_node = root.find('size')
        if filename_node is None or size_node is None:
            # print(f"Skipping {xml_file.name}: Missing filename or size tag.") # Less verbose
            return None, None

        filename = filename_node.text
        width_node = size_node.find('width')
        height_node = size_node.find('height')

        if width_node is None or height_node is None:
            # print(f"Skipping {xml_file.name}: Missing width or height tag.") # Less verbose
            return None, None

        width = int(width_node.text)
        height = int(height_node.text)
        image_path = find_image_path(filename, base_img_dir) # Pass the correct base image dir

        if not image_path:
             # print(f"Skipping {xml_file.name}: Cannot find image file {filename} in {base_img_dir}.") # Less verbose
             return None, None


        annotations = []
        objects_found = root.findall('object')
        # Keep image info even if no objects found (for potential background images)

        for obj in objects_found:
            name_node = obj.find('name')
            bndbox_node = obj.find('bndbox')
            if name_node is None or bndbox_node is None:
                # print(f"Warning: Skipping object in {xml_file.name}: Missing name or bndbox tag.") # Less verbose
                continue

            name = name_node.text.lower().replace(" ", "_")
            if name not in CLASS_MAP:
                continue

            xmin_node = bndbox_node.find('xmin')
            ymin_node = bndbox_node.find('ymin')
            xmax_node = bndbox_node.find('xmax')
            ymax_node = bndbox_node.find('ymax')

            if None in [xmin_node, ymin_node, xmax_node, ymax_node]:
                 # print(f"Warning: Skipping object '{name}' in {xml_file.name}: Missing bbox coordinate tags.") # Less verbose
                 continue

            try:
                xmin = int(float(xmin_node.text))
                ymin = int(float(ymin_node.text))
                xmax = int(float(xmax_node.text))
                ymax = int(float(ymax_node.text))
            except ValueError:
                 # print(f"Warning: Skipping object '{name}' in {xml_file.name}: Non-numeric bbox coordinates.") # Less verbose
                 continue

            if xmin >= xmax or ymin >= ymax or xmin < 0 or ymin < 0 or xmax > width or ymax > height:
                # print(f"Warning: Invalid bbox {name} in {xml_file.name}: ... Skipping object.") # Less verbose
                continue

            annotations.append({
                'class': name, 'class_id': CLASS_MAP[name],
                'xmin': xmin, 'ymin': ymin, 'xmax': xmax, 'ymax': ymax
            })

        image_info = {
            'filename': filename, 'width': width, 'height': height,
            'image_path': str(image_path), 'annotations': annotations
        }
        return filename, image_info

    except ET.ParseError: # print(f"Error parsing XML file: {xml_file}"); # Less verbose
                          return None, None
    except FileNotFoundError: # print(f"Error: XML file not found: {xml_file}"); # Less verbose
                             return None, None
    except Exception as e: # print(f"Unexpected error processing {xml_file}: {e}"); # Less verbose
                          return None, None

def find_image_path(filename, base_img_dir):
    """Finds the full path of an image file within subdirectories of the base_img_dir."""
    extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff']
    base_filename = Path(filename).stem

    # Check common locations first
    for ext in extensions:
        potential_path = base_img_dir / (base_filename + ext)
        if potential_path.is_file(): return potential_path
        for subdir in base_img_dir.iterdir():
            if subdir.is_dir():
                potential_path_subdir = subdir / (base_filename + ext)
                if potential_path_subdir.is_file(): return potential_path_subdir

    # Fallback search if filename in XML might have extension already or different case
    for root, _, files in os.walk(base_img_dir):
         # Case-insensitive check might be useful depending on dataset consistency
         for f in files:
             if f.lower() == filename.lower():
                 return Path(root) / f
    return None # Not found

def load_all_annotations(annot_dir, img_dir):
    """Loads all valid annotations from the XML files."""
    all_image_data = {}
    xml_files = list(annot_dir.rglob("*.xml"))
    print(f"Found {len(xml_files)} XML files in {annot_dir}.")
    if not xml_files:
        print(f"Warning: No XML files found. Check 'ANNOTATIONS_DIR': {annot_dir}")
        return {}

    processed_count = 0
    skipped_count = 0
    files_with_no_objects = 0
    for xml_file in tqdm(xml_files, desc="Parsing XMLs"):
        filename, image_info = parse_xml(xml_file, img_dir)
        if filename and image_info:
            if filename not in all_image_data:
                 all_image_data[filename] = image_info
                 processed_count += 1
                 if not image_info['annotations']:
                     files_with_no_objects += 1
        else:
            skipped_count += 1

    print(f"Processed {processed_count} annotation files.")
    if files_with_no_objects > 0: print(f"  ({files_with_no_objects} had no object annotations)")
    if skipped_count > 0: print(f"Skipped {skipped_count} files (parse error, invalid data, or image missing).")
    print(f"Final dataset size (images with parsed info): {len(all_image_data)}")
    return all_image_data
# <<< END of unchanged functions >>>

# --- Load the data ---
print("Loading annotations...")
all_data = load_all_annotations(ANNOTATIONS_DIR, IMAGES_DIR)
all_image_ids = list(all_data.keys())

if not all_image_ids:
    raise ValueError("No data loaded. Check dataset paths and XML parsing.")

# --- Split Data (Using your fixed version) ---
test_frac = 0.15
val_frac = 0.15
train_ids, test_ids = train_test_split(all_image_ids, test_size=test_frac, random_state=SEED)
if len(train_ids) < 1: # Need at least one sample for training
    raise ValueError("Not enough data left for training after test split.")
relative_val_frac = val_frac / (1.0 - test_frac) if (1.0 - test_frac) > 0 else 0.5
if relative_val_frac >= 1.0 or len(train_ids) < 2:
    val_ids = [] # No validation set if remaining data is too small or fraction is too high
else:
    train_ids, val_ids = train_test_split(train_ids, test_size=relative_val_frac, random_state=SEED)

print(f"Dataset split: Train={len(train_ids)}, Validation={len(val_ids)}, Test={len(test_ids)}")

# --- Visualization Helper (Unchanged from your fixed version) ---
# ... (Keep the fixed visualize_bbox function here) ...
def visualize_bbox(image_path_str, annotations, model_preds=None, title="Annotations"):
    """Draws ground truth and optional prediction bounding boxes on an image."""
    image_path = Path(image_path_str) # Ensure it's a Path object
    if not image_path.is_file():
        print(f"Error: Image file not found at {image_path}")
        return

    image = cv2.imread(str(image_path))
    if image is None:
        print(f"Error reading image with OpenCV: {image_path}")
        return
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Draw Ground Truth
    for annot in annotations:
        try:
            xmin, ymin, xmax, ymax = int(annot['xmin']), int(annot['ymin']), int(annot['xmax']), int(annot['ymax'])
            label = annot['class']
            cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2) # Green for GT
            text_y = ymin - 10 if ymin > 20 else ymin + 20
            cv2.putText(image, f"GT: {label}", (xmin, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
        except KeyError as e: print(f"Warning: Missing key {e} in GT annotation: {annot}")
        except Exception as e: print(f"Warning: Error drawing GT box for {annot}: {e}")

    # Draw Predictions
    if model_preds:
        for pred in model_preds:
            try:
                xmin, ymin, xmax, ymax = int(pred['box'][0]), int(pred['box'][1]), int(pred['box'][2]), int(pred['box'][3])
                label = INV_CLASS_MAP.get(pred['label'], f"Unknown:{pred['label']}")
                score = pred['score']
                cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (255, 0, 0), 2) # Red for Pred
                text_y = ymin - 25 if ymin > 40 else ymin + 40
                cv2.putText(image, f"P: {label} ({score:.2f})", (xmin, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
            except KeyError as e: print(f"Warning: Missing key {e} in prediction: {pred}")
            except Exception as e: print(f"Warning: Error drawing prediction box for {pred}: {e}")

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
# <<< END of unchanged visualization function >>>

# ==============================================================================
# 2. YOLOv8 Implementation (Ultralytics)
# ==============================================================================
print("\n--- Preparing Data for YOLOv8 ---")

# <<< Function prepare_yolo_data unchanged >>>
def prepare_yolo_data(image_ids, dataset_split, target_dir, source_data, img_size):
    """Copies images and creates YOLO format label files."""
    img_path = target_dir / 'images' / dataset_split
    lbl_path = target_dir / 'labels' / dataset_split
    img_path.mkdir(parents=True, exist_ok=True)
    lbl_path.mkdir(parents=True, exist_ok=True)
    copied_count = 0
    label_count = 0
    for img_id in tqdm(image_ids, desc=f"Preparing YOLO {dataset_split}"):
        if img_id not in source_data: continue # Skip if ID somehow missing
        data = source_data[img_id]
        source_img_path = Path(data['image_path'])
        target_img_path = img_path / source_img_path.name
        try:
            shutil.copyfile(source_img_path, target_img_path)
            copied_count +=1
        except Exception as e:
            print(f"Error copying {source_img_path} to {target_img_path}: {e}")
            continue
        width, height = data['width'], data['height']
        label_file_path = lbl_path / f"{source_img_path.stem}.txt"
        with open(label_file_path, 'w') as f:
            for annot in data['annotations']:
                class_id = annot['class_id']
                xmin, ymin, xmax, ymax = annot['xmin'], annot['ymin'], annot['xmax'], annot['ymax']
                x_center = ((xmin + xmax) / 2) / width
                y_center = ((ymin + ymax) / 2) / height
                bbox_width = (xmax - xmin) / width
                bbox_height = (ymax - ymin) / height
                x_center, y_center = max(0.0, min(1.0, x_center)), max(0.0, min(1.0, y_center))
                bbox_width, bbox_height = max(0.0, min(1.0, bbox_width)), max(0.0, min(1.0, bbox_height))
                f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {bbox_width:.6f} {bbox_height:.6f}\n")
                label_count += 1
    print(f"Prepared {copied_count} images and {label_count} labels for {dataset_split}.")
# <<< END of unchanged prepare_yolo_data >>>

# Prepare data splits for YOLO
prepare_yolo_data(train_ids, 'train', YOLO_DATA_DIR, all_data, IMG_SIZE_YOLO)
if val_ids: prepare_yolo_data(val_ids, 'val', YOLO_DATA_DIR, all_data, IMG_SIZE_YOLO)
if test_ids: prepare_yolo_data(test_ids, 'test', YOLO_DATA_DIR, all_data, IMG_SIZE_YOLO)

# --- Create data.yaml for YOLO ---
data_yaml_content = {
    'path': str(YOLO_DATA_DIR.resolve()), # Absolute path
    'train': 'images/train',
    'val': 'images/val' if val_ids else None, # Handle case with no validation set
    'test': 'images/test' if test_ids else None, # Handle case with no test set
    'names': {i: name for i, name in enumerate(CLASSES)}
}

yolo_yaml_path = YOLO_DATA_DIR / 'data.yaml'
with open(yolo_yaml_path, 'w') as f:
    yaml.dump(data_yaml_content, f, default_flow_style=False, sort_keys=False)

print(f"YOLO data YAML created at: {yolo_yaml_path}")

# --- Train YOLOv8 ---
print("\n--- Training YOLOv8 ---")
# Define project directory within /kaggle/working/ for outputs
yolo_project_dir = OUTPUT_DIR / 'yolo_training_output' # Changed from MODEL_CHECKPOINTS_DIR to avoid nesting clash
yolo_run_name = 'yolov8s_pcb_run1'

yolo_model = YOLO('yolov8s.pt') # Load pre-trained weights

# Check if validation set exists for early stopping and saving best model
use_validation = bool(val_ids)
print(f"YOLO Training with validation set: {use_validation}")

# Adjust hyperparameters as needed
yolo_results = yolo_model.train(
    data=str(yolo_yaml_path),
    epochs=EPOCHS_YOLO,
    imgsz=IMG_SIZE_YOLO,
    batch=BATCH_SIZE_YOLO,
    project=str(yolo_project_dir), # Save main results here (/kaggle/working/yolo_training_output)
    name=yolo_run_name,           # Subdirectory name (/kaggle/working/yolo_training_output/yolov8s_pcb_run1)
    exist_ok=True,
    patience=10 if use_validation else 0, # Use patience only if validation set exists
    seed=SEED,
    save=True, # Explicitly ensure saving is enabled
    save_period=-1, # Default: only save last and best
    val=use_validation, # Explicitly enable/disable validation steps
)

print("YOLOv8 Training complete.")

# --- Identify and Copy Best YOLOv8 Model ---
# Ultralytics saves best.pt in the run directory based on validation metrics
yolo_run_dir = yolo_project_dir / yolo_run_name
yolo_weights_dir = yolo_run_dir / 'weights'
potential_best_yolo_path = yolo_weights_dir / 'best.pt'

# Define final save path in our designated checkpoints directory
final_best_yolo_model_path = MODEL_CHECKPOINTS_DIR / 'yolov8s_best.pt' # Simple name in checkpoints dir

if potential_best_yolo_path.exists():
    print(f"Found best YOLOv8 model at: {potential_best_yolo_path}")
    try:
        shutil.copyfile(potential_best_yolo_path, final_best_yolo_model_path)
        print(f"Copied best YOLOv8 model for download to: {final_best_yolo_model_path}")
        best_yolo_model_path_for_eval = final_best_yolo_model_path # Use this path for evaluation
    except Exception as e:
        print(f"Error copying best YOLO model: {e}")
        print("Evaluation will use the path within the run directory.")
        best_yolo_model_path_for_eval = potential_best_yolo_path # Fallback
else:
    print(f"Warning: best.pt not found in {yolo_weights_dir}.")
    last_yolo_path = yolo_weights_dir / 'last.pt'
    if last_yolo_path.exists():
         print("Using last.pt instead for evaluation and saving.")
         try:
             final_last_yolo_model_path = MODEL_CHECKPOINTS_DIR / 'yolov8s_last.pt'
             shutil.copyfile(last_yolo_path, final_last_yolo_model_path)
             print(f"Copied last YOLOv8 model for download to: {final_last_yolo_model_path}")
             best_yolo_model_path_for_eval = final_last_yolo_model_path # Use this path
         except Exception as e:
             print(f"Error copying last YOLO model: {e}")
             best_yolo_model_path_for_eval = last_yolo_path # Fallback
    else:
        print("Neither best.pt nor last.pt found. Cannot evaluate or save final YOLO model.")
        best_yolo_model_path_for_eval = None


# --- Evaluate YOLOv8 (using the identified best/last model) ---
if best_yolo_model_path_for_eval and best_yolo_model_path_for_eval.exists():
    print("\n--- Evaluating YOLOv8 on Test Set ---")
    yolo_model_eval = YOLO(str(best_yolo_model_path_for_eval))

    if test_ids: # Only evaluate if test set exists
        yolo_metrics = yolo_model_eval.val(split='test')
        print("YOLOv8 Test Metrics:")
        try:
            print(f"  mAP50-95: {yolo_metrics.box.map:.4f}")
            print(f"  mAP50:    {yolo_metrics.box.map50:.4f}")
            print(f"  mAP75:    {yolo_metrics.box.map75:.4f}")
        except Exception as e:
            print(f"Could not print metrics directly. Error: {e}")
            # print(yolo_metrics) # Print full object if needed
    else:
        print("Skipping YOLOv8 test set evaluation as no test split was defined.")

    # --- Predict with YOLOv8 (Example) ---
    # ... (Prediction code remains the same, uses yolo_model_eval) ...
    if test_ids:
        print("\n--- Running YOLOv8 Prediction (Example) ---")
        yolo_test_img_dir = YOLO_DATA_DIR / 'images' / 'test'
        if list(yolo_test_img_dir.glob("*.jpg")): # Check if images exist
            sample_test_img_path = random.choice(list(yolo_test_img_dir.glob("*.jpg")))
            pred_results = yolo_model_eval.predict(source=str(sample_test_img_path), conf=0.25)
            img_pred = pred_results[0].plot()
            img_pred_rgb = cv2.cvtColor(img_pred, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(10, 10)); plt.imshow(img_pred_rgb); plt.title(f"YOLOv8 Pred: {sample_test_img_path.name}"); plt.axis('off'); plt.show()
        else:
             print("No images found in YOLO test directory for prediction example.")

else:
    print(f"Could not find a trained YOLO model to evaluate. Skipping evaluation.")

# ==============================================================================
# 3. Faster R-CNN Implementation (TorchVision)
# ==============================================================================
print("\n--- Preparing Data and Training Faster R-CNN ---")

# <<< Ensure torchvision transforms are imported correctly >>>
# Using v2 style from previous version
from torchvision.transforms.v2 import functional as F, ConvertImageDtype, Compose, RandomHorizontalFlip
# <<< END Ensure torchvision transforms are imported correctly >>>


# --- PyTorch Dataset (Unchanged from your fixed version) ---
# ... (Keep the fixed PCBDataset class here) ...
class PCBDataset(Dataset):
    def __init__(self, image_ids, data_dict, transforms=None):
        super().__init__()
        self.image_ids = image_ids
        self.data = data_dict
        self.transforms = transforms

    def __len__(self): return len(self.image_ids)

    def __getitem__(self, index):
        image_id = self.image_ids[index]
        if image_id not in self.data:
            print(f"Error: Image ID {image_id} not found in data dictionary.")
            # Return None or raise error, or return dummy data
            # Returning dummy data might hide issues but allows continuation
            # Let's return None and handle it in collate_fn or DataLoader if needed
            return None, None # Or raise IndexError(f"Image ID {image_id} not found")

        record = self.data[image_id]
        image_path = record.get('image_path')
        if not image_path or not Path(image_path).is_file():
             print(f"Error: Image path missing or invalid for ID {image_id}: {image_path}")
             return None, None

        image = cv2.imread(image_path)
        if image is None:
            print(f"Error reading image: {image_path}")
            return None, None
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        boxes, labels, areas, iscrowd = [], [], [], []
        for annot in record['annotations']:
            xmin, ymin, xmax, ymax = annot['xmin'], annot['ymin'], annot['xmax'], annot['ymax']
            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(annot['class_id'] + 1) # Background = 0
            areas.append((xmax - xmin) * (ymax - ymin))
            iscrowd.append(0)

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        # Handle no annotations case: target requires boxes/labels even if empty
        if not record['annotations']:
            target = {
                "boxes": torch.zeros((0, 4), dtype=torch.float32),
                "labels": torch.zeros(0, dtype=torch.int64),
                "image_id": torch.tensor([index]),
                "area": torch.zeros(0, dtype=torch.float32),
                "iscrowd": torch.zeros(0, dtype=torch.int64)
            }
        else:
            target = {
                "boxes": boxes,
                "labels": torch.as_tensor(labels, dtype=torch.int64),
                "image_id": torch.tensor([index]),
                "area": torch.as_tensor(areas, dtype=torch.float32),
                "iscrowd": torch.as_tensor(iscrowd, dtype=torch.int64)
            }

        if self.transforms:
             try:
                 image = F.to_image(image) # Convert numpy array (HWC) to tensor (CHW) [0, 255] uint8
                 # Only pass target to transforms if it contains boxes
                 if target['boxes'].shape[0] > 0:
                     image, target = self.transforms(image, target)
                 else:
                     # Apply transforms that work on image only (like type conversion)
                     # Need to handle which transforms expect target
                     # Simplification: assume ToTensor/ConvertImageDtype is always last
                     temp_transforms = Compose([t for t in self.transforms.transforms if not isinstance(t, RandomHorizontalFlip)]) # Example: exclude transforms needing target
                     image = temp_transforms(image)

                 # Ensure image dtype is float32 after transforms
                 image = image.to(torch.float32)
                 # Ensure target boxes are float32
                 if "boxes" in target: target["boxes"] = target["boxes"].to(torch.float32)

             except Exception as e:
                 print(f"Error during transform for image index {index} (ID: {image_id}): {e}")
                 return None, None # Skip item if transform fails


             # Safety check after transform
             if "boxes" in target and target["boxes"].shape[0] > 0:
                valid_boxes_idx = (target["boxes"][:, 2] > target["boxes"][:, 0]) & (target["boxes"][:, 3] > target["boxes"][:, 1])
                if not torch.all(valid_boxes_idx):
                    target["boxes"] = target["boxes"][valid_boxes_idx]
                    target["labels"] = target["labels"][valid_boxes_idx]
                    target["area"] = target["area"][valid_boxes_idx]
                    target["iscrowd"] = target["iscrowd"][valid_boxes_idx]
                    if target["boxes"].shape[0] == 0:
                         del target["boxes"]; del target["labels"]; del target["area"]; del target["iscrowd"]

        # Convert image to float32 before returning if not done by transforms
        if not isinstance(image, torch.Tensor) or image.dtype != torch.float32:
             image = F.to_image(image) # To Tensor if still numpy
             image = F.convert_image_dtype(image, torch.float32) # To float

        return image, target
# <<< END of unchanged PCBDataset >>>


# --- Transforms (Using Torchvision v2 - unchanged) ---
def get_transform(train):
    transforms = []
    if train:
        transforms.append(RandomHorizontalFlip(p=0.5))
    # ConvertImageDtype handles scaling to [0,1] and dtype conversion
    transforms.append(ConvertImageDtype(torch.float))
    return Compose(transforms)

# --- Datasets and DataLoaders ---
# Handle cases where splits might be empty
dataset_train_rcnn = PCBDataset(train_ids, all_data, transforms=get_transform(train=True)) if train_ids else None
dataset_val_rcnn = PCBDataset(val_ids, all_data, transforms=get_transform(train=False)) if val_ids else None
dataset_test_rcnn = PCBDataset(test_ids, all_data, transforms=get_transform(train=False)) if test_ids else None

# Collate function to handle potential None values from dataset errors
def collate_fn(batch):
    # Filter out None items first
    batch = [item for item in batch if item is not None and item[0] is not None and item[1] is not None]
    if not batch: # Return None or empty tuple if batch becomes empty after filtering
        return None, None
    # Proceed with default collation
    return tuple(zip(*batch))

# Create DataLoaders only if datasets exist
data_loader_train_rcnn = DataLoader(
    dataset_train_rcnn, batch_size=BATCH_SIZE_RCNN, shuffle=True, num_workers=2,
    collate_fn=collate_fn, pin_memory=True if DEVICE == 'cuda' else False
) if dataset_train_rcnn else None

data_loader_val_rcnn = DataLoader(
    dataset_val_rcnn, batch_size=BATCH_SIZE_RCNN, shuffle=False, num_workers=2,
    collate_fn=collate_fn, pin_memory=True if DEVICE == 'cuda' else False
) if dataset_val_rcnn else None

data_loader_test_rcnn = DataLoader(
    dataset_test_rcnn, batch_size=BATCH_SIZE_RCNN, shuffle=False, num_workers=2,
    collate_fn=collate_fn, pin_memory=True if DEVICE == 'cuda' else False
) if dataset_test_rcnn else None


# --- Define Faster R-CNN Model (Unchanged) ---
# ... (Keep get_faster_rcnn_model function) ...
def get_faster_rcnn_model(num_classes):
    model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights='DEFAULT')
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes + 1) # +1 for background
    return model

# --- Instantiate Model, Optimizer, Scheduler (Check for valid train loader) ---
rcnn_model = None
optimizer_rcnn = None
lr_scheduler_rcnn = None

if data_loader_train_rcnn:
    rcnn_model = get_faster_rcnn_model(NUM_CLASSES)
    rcnn_model.to(DEVICE)
    params = [p for p in rcnn_model.parameters() if p.requires_grad]
    optimizer_rcnn = torch.optim.SGD(params, lr=LEARNING_RATE_RCNN, momentum=0.9, weight_decay=0.0005)
    lr_scheduler_rcnn = torch.optim.lr_scheduler.StepLR(optimizer_rcnn, step_size=8, gamma=0.1)
else:
    print("Skipping Faster R-CNN model setup: No training data available.")


# --- Training Loop (Unchanged) ---
# ... (Keep train_one_epoch function) ...
def train_one_epoch(model, optimizer, data_loader, device, epoch, total_epochs):
    model.train()
    total_loss = 0
    progress_bar = tqdm(data_loader, desc=f"Epoch {epoch+1}/{total_epochs} [Training]")
    batches_processed = 0
    for i, batch_data in enumerate(progress_bar):
        # Handle collation result if it filters the whole batch
        if batch_data is None or not batch_data[0]:
             # print(f"Skipping empty or invalid batch {i}") # Less verbose
             continue

        images, targets = batch_data
        images = list(image.to(device) for image in images)
        targets = [{k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in t.items()} for t in targets]

        valid_targets = []
        valid_images = []
        for img, t in zip(images, targets):
            if "boxes" in t and t['boxes'].shape[0] > 0:
                 valid_targets.append(t)
                 valid_images.append(img)

        if not valid_images: continue # Skip batch if no images had valid targets

        try:
            loss_dict = model(valid_images, valid_targets)
            losses = sum(loss for loss in loss_dict.values())
            loss_value = losses.item()

            if torch.isnan(losses) or torch.isinf(losses):
                 print(f"Warning: NaN or Inf loss detected at batch {i}. Skipping batch.")
                 # Consider stopping or reducing LR if this happens frequently
                 continue

            total_loss += loss_value
            batches_processed += 1

            optimizer.zero_grad()
            losses.backward()
            # Optional: Gradient clipping
            # torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            if batches_processed > 0:
                 progress_bar.set_postfix(loss=f"{loss_value:.4f}", avg_loss=f"{total_loss / batches_processed:.4f}")

        except Exception as e:
             print(f"\nError during training batch {i}: {e}")
             # Optionally continue to next batch or stop
             continue # Continue to next batch

    return total_loss / batches_processed if batches_processed > 0 else 0.0

# --- Evaluation Loop (Unchanged) ---
# ... (Keep evaluate function) ...
@torch.inference_mode()
def evaluate(model, data_loader, device):
    model.eval()
    if not MeanAveragePrecision: return {}

    metric = MeanAveragePrecision(box_format='xyxy', iou_type='bbox').to(device)
    progress_bar = tqdm(data_loader, desc="Evaluation")

    for batch_data in progress_bar:
        if batch_data is None or not batch_data[0]: continue # Skip empty batches
        images, targets = batch_data
        images = list(image.to(device) for image in images)
        targets = [{k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in t.items()} for t in targets]

        try:
             outputs = model(images)
        except Exception as e:
             print(f"\nError during evaluation inference: {e}")
             continue # Skip batch on error

        # Format for torchmetrics
        formatted_targets = []
        valid_indices = [] # Track indices of targets that are valid for evaluation
        for idx, t in enumerate(targets):
            if "boxes" in t and t['boxes'].shape[0] > 0:
                formatted_targets.append({"boxes": t["boxes"], "labels": t["labels"]})
                valid_indices.append(idx)
            # else: # Handle empty ground truth - torchmetrics should handle this
            #     formatted_targets.append({"boxes": torch.empty((0, 4), device=device), "labels": torch.empty(0, device=device, dtype=torch.int64)})

        # Filter outputs to match valid targets if necessary, although torchmetrics might handle alignment
        # Let's try passing all outputs and see if torchmetrics handles it correctly first
        # Ensure outputs indices align with targets (should be the case if no filtering done here)
        if len(outputs) != len(formatted_targets):
             print(f"Warning: Mismatch between number of outputs ({len(outputs)}) and formatted targets ({len(formatted_targets)}). Check filtering.")
             # Use safer filtering based on valid_indices if mismatch occurs
             filtered_outputs = [outputs[i] for i in valid_indices]
             if not filtered_outputs: continue # Skip if no valid preds/targets remain
             metric.update(filtered_outputs, formatted_targets) # Update with filtered lists
        else:
             # If lengths match, assume alignment and update directly
             metric.update(outputs, targets) # Pass original targets (dicts with boxes/labels on device)


    try:
        results = metric.compute()
        return results
    except Exception as e:
        print(f"Error computing metrics: {e}")
        return {}

# --- Run Faster R-CNN Training (with checks) ---
best_val_map = -1.0
rcnn_history = {'train_loss': [], 'val_map': [], 'val_map_50': []}
# <<< Define final save path outside the loop >>>
best_rcnn_model_path = MODEL_CHECKPOINTS_DIR / 'faster_rcnn_best.pth'

if rcnn_model and optimizer_rcnn and data_loader_train_rcnn:
    print("Starting Faster R-CNN training...")
    use_rcnn_validation = bool(data_loader_val_rcnn)
    print(f"Faster R-CNN Training with validation set: {use_rcnn_validation}")

    for epoch in range(EPOCHS_RCNN):
        avg_train_loss = train_one_epoch(rcnn_model, optimizer_rcnn, data_loader_train_rcnn, DEVICE, epoch, EPOCHS_RCNN)
        rcnn_history['train_loss'].append(avg_train_loss)
        print_prefix = f"Epoch {epoch+1}/{EPOCHS_RCNN} - Train Loss: {avg_train_loss:.4f}"

        val_map = -1.0 # Default if no validation
        val_map_50 = -1.0

        # Validation step (only if validation loader exists)
        if use_rcnn_validation:
            val_metrics = evaluate(rcnn_model, data_loader_val_rcnn, DEVICE)
            val_map = val_metrics.get('map', torch.tensor(-1.0)).item()
            val_map_50 = val_metrics.get('map_50', torch.tensor(-1.0)).item()
            rcnn_history['val_map'].append(val_map)
            rcnn_history['val_map_50'].append(val_map_50)
            print(f"{print_prefix} - Val mAP: {val_map:.4f} - Val mAP@.50: {val_map_50:.4f}")

            # Save the best model based on validation mAP
            if val_map > best_val_map:
                best_val_map = val_map
                try:
                    torch.save(rcnn_model.state_dict(), best_rcnn_model_path)
                    print(f"Saved **best** Faster R-CNN model state_dict to: {best_rcnn_model_path} (mAP: {best_val_map:.4f})")
                except Exception as e:
                    print(f"Error saving best Faster R-CNN model: {e}")
        else:
             print(f"{print_prefix} - (No validation set)")
             # If no validation, save the model from the last epoch
             if epoch == EPOCHS_RCNN - 1:
                  last_rcnn_model_path = MODEL_CHECKPOINTS_DIR / 'faster_rcnn_last.pth'
                  try:
                       torch.save(rcnn_model.state_dict(), last_rcnn_model_path)
                       print(f"Saved **last** Faster R-CNN model state_dict to: {last_rcnn_model_path} (No validation performed)")
                       # Use last model path if best doesn't exist
                       if not best_rcnn_model_path.exists(): best_rcnn_model_path = last_rcnn_model_path
                  except Exception as e:
                      print(f"Error saving last Faster R-CNN model: {e}")


        # Update the learning rate scheduler
        if lr_scheduler_rcnn:
            lr_scheduler_rcnn.step()

    print("Faster R-CNN Training complete.")
    # <<< Add confirmation print AFTER the loop >>>
    if best_rcnn_model_path.exists():
         print(f"\nFinal best Faster R-CNN model state_dict saved at: {best_rcnn_model_path}")
    else:
         print("\nWarning: No best Faster R-CNN model was saved (check validation performance or saving logic).")

else:
    print("Skipping Faster R-CNN training.")


# --- Evaluate Faster R-CNN on Test Set ---
print("\n--- Evaluating Faster R-CNN on Test Set ---")
# Use the determined best_rcnn_model_path (could be best or last)
if 'best_rcnn_model_path' in locals() and best_rcnn_model_path.exists() and data_loader_test_rcnn:
    print(f"Loading model from: {best_rcnn_model_path}")
    rcnn_model_eval = get_faster_rcnn_model(NUM_CLASSES)
    try:
        rcnn_model_eval.load_state_dict(torch.load(best_rcnn_model_path, map_location=DEVICE))
        rcnn_model_eval.to(DEVICE)
        rcnn_model_eval.eval()

        test_metrics = evaluate(rcnn_model_eval, data_loader_test_rcnn, DEVICE)
        print("Faster R-CNN Test Metrics:")
        for key, value in test_metrics.items():
             if isinstance(value, torch.Tensor) and value.numel() == 1: print(f"  {key}: {value.item():.4f}")
             else: print(f"  {key}: {value}") # Handle non-scalar tensors/other types

        # --- Predict with Faster R-CNN (Example) ---
        # ... (Prediction code remains the same, uses rcnn_model_eval) ...
        if test_ids and dataset_test_rcnn: # Check if test data exists
            print("\n--- Running Faster R-CNN Prediction (Example) ---")
            try:
                 random_idx = random.randint(0, len(dataset_test_rcnn)-1)
                 img_tensor, target = dataset_test_rcnn[random_idx] # Get raw tensor and target
                 if img_tensor is None or target is None: raise ValueError("Dataset returned None")

                 # Find original path carefully using image_id stored in target
                 original_image_id_index = target['image_id'].item()
                 original_image_filename = dataset_test_rcnn.image_ids[original_image_id_index]
                 original_image_path = dataset_test_rcnn.data[original_image_filename]['image_path']
                 gt_annotations = dataset_test_rcnn.data[original_image_filename]['annotations']

                 with torch.no_grad():
                     prediction = rcnn_model_eval([img_tensor.to(DEVICE)])[0]

                 pred_boxes = prediction['boxes'].cpu().numpy()
                 pred_labels = prediction['labels'].cpu().numpy()
                 pred_scores = prediction['scores'].cpu().numpy()
                 score_threshold = 0.3
                 keep_indices = pred_scores > score_threshold
                 pred_boxes, pred_labels, pred_scores = pred_boxes[keep_indices], pred_labels[keep_indices], pred_scores[keep_indices]
                 valid_label_indices = pred_labels > 0 # Exclude background
                 pred_boxes, pred_labels, pred_scores = pred_boxes[valid_label_indices], pred_labels[valid_label_indices] - 1, pred_scores[valid_label_indices]

                 formatted_preds = [{'box': box, 'label': label, 'score': score} for box, label, score in zip(pred_boxes, pred_labels, pred_scores)]
                 visualize_bbox(original_image_path, gt_annotations, formatted_preds, title=f"Faster R-CNN Pred: {Path(original_image_path).name}")

            except IndexError: print("Error: Invalid index generated for prediction sample.")
            except KeyError: print("Error: Key not found when retrieving image path/annotations for prediction.")
            except ValueError as e: print(f"Error during prediction sample retrieval or processing: {e}")
            except Exception as e: print(f"An unexpected error occurred during Faster R-CNN prediction visualization: {e}")

    except Exception as e:
        print(f"Error loading Faster R-CNN model state_dict from {best_rcnn_model_path}: {e}")

elif not ('best_rcnn_model_path' in locals() and best_rcnn_model_path.exists()):
    print("Could not find saved Faster R-CNN model. Skipping evaluation.")
elif not data_loader_test_rcnn:
    print("No test data loader available for Faster R-CNN. Skipping evaluation.")


# ==============================================================================
# 4. EfficientDet (TF Hub Inference Only - No Model Saved Here)
# ==============================================================================
print("\n--- EfficientDet (TF Hub Inference - No Training/Saving in this script) ---")
# <<< Add comment here clarifying no model is saved from this section >>>
# Note: This section only performs inference using a pre-trained model from TensorFlow Hub.
# No model weights generated by *this script* will be saved for EfficientDet.
# To use EfficientDet in an application, you would typically load it directly from TF Hub again.

# --- Example: Loading and running inference with TF Hub EfficientDet ---
effdet_inference_time = -1.0
physical_devices_gpu = [] # Default value

try:
    # <<< Ensure TF imports are active if running this section >>>
    import tensorflow as tf
    import tensorflow_hub as hub
    # <<< END Ensure TF imports >>>

    print(f"TensorFlow Version: {tf.__version__}")
    print(f"TF Hub Version: {hub.__version__}")
    physical_devices_gpu = tf.config.list_physical_devices('GPU')
    target_device = '/CPU:0' # Force CPU for TF Hub model to potentially avoid cuDNN issues on Kaggle
    print(f"Attempting to run TF Hub model on CPU using tf.device('{target_device}') context.")

    with tf.device(target_device):
        hub_url = "https://tfhub.dev/tensorflow/efficientdet/d0/1"
        print(f"Loading EfficientDet model from TF Hub: {hub_url}")
        detector = hub.load(hub_url)
        print("EfficientDet model loaded.")

        if not test_ids: print("Warning: No test images available for EfficientDet inference.")
        else:
            # ... (rest of the TF Hub inference and visualization code - unchanged) ...
            sample_id_for_effdet = random.choice(test_ids)
            if sample_id_for_effdet not in all_data: print(f"Error: Sample ID {sample_id_for_effdet} not found in all_data.")
            else:
                effdet_img_path = all_data[sample_id_for_effdet]['image_path']
                if not Path(effdet_img_path).is_file(): print(f"Error: Image file not found: {effdet_img_path}")
                else:
                    img_np = cv2.imread(effdet_img_path)
                    if img_np is None: print(f"Error: Failed to read image {effdet_img_path}")
                    else:
                        img_np = cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB)
                        if img_np.ndim == 3 and img_np.shape[2] == 3:
                            img_tf = tf.convert_to_tensor(img_np, dtype=tf.uint8)[tf.newaxis, ...]
                            print(f"Running EfficientDet inference on: {effdet_img_path} using device: {target_device}")
                            start_time = time.time()
                            detector_output = detector(img_tf)
                            end_time = time.time()
                            effdet_inference_time = end_time - start_time
                            print(f"EfficientDet Inference Time: {effdet_inference_time:.4f} seconds")

                            boxes = detector_output["detection_boxes"][0].numpy()
                            scores = detector_output["detection_scores"][0].numpy()
                            classes = detector_output["detection_classes"][0].numpy().astype(int)
                            num_detections = int(detector_output["num_detections"][0])

                            h, w, _ = img_np.shape
                            effdet_preds_viz = []
                            score_thresh_effdet = 0.3
                            for i in range(num_detections):
                                if scores[i] >= score_thresh_effdet:
                                    ymin, xmin, ymax, xmax = boxes[i]
                                    box_denorm = [max(0, xmin * w), max(0, ymin * h), min(w - 1, xmax * w), min(h - 1, ymax * h)]
                                    label = f"TFHub Cls: {classes[i]}" # COCO classes, not PCB
                                    generic_pcb_label_index = 0
                                    effdet_preds_viz.append({'box': box_denorm,'label': generic_pcb_label_index,'score': scores[i]})

                            print(f"Found {len(effdet_preds_viz)} EfficientDet detections above threshold {score_thresh_effdet}")
                            gt_annotations_effdet = all_data[sample_id_for_effdet]['annotations']
                            visualize_bbox(effdet_img_path, gt_annotations_effdet, effdet_preds_viz, title=f"EfficientDet (TF Hub) Pred: {Path(effdet_img_path).name}")
                        else: print(f"Error: Invalid image dimensions: {img_np.shape}")

except ImportError: print("\nTensorFlow/TF Hub not installed. Skipping EfficientDet.")
except Exception as e:
    import traceback
    print(f"\nError during EfficientDet TF Hub execution: {e}")
    # traceback.print_exc() # Uncomment for detailed traceback


# ==============================================================================
# 5. Comparison Summary & Download Instructions
# ==============================================================================
print("\n--- Comparison Summary & Download Instructions ---")

# --- Inference Time Measurement Function (Unchanged from your fixed version) ---
# ... (Keep the fixed measure_inference function here) ...
def measure_inference(model, data_loader, device, model_type='torchvision'):
    model.eval()
    times = []
    num_samples = 0
    warmup_batches = 5
    if isinstance(device, str): pytorch_device = torch.device(device)
    elif isinstance(device, torch.device): pytorch_device = device
    else: pytorch_device = torch.device('cpu')
    try: model.to(pytorch_device)
    except Exception as e: print(f"Warning: Cannot move model to {pytorch_device}: {e}"); # return -1

    total_batches = len(data_loader) if hasattr(data_loader, '__len__') else 0
    if total_batches == 0: print(f"Warning: Dataloader {model_type} empty."); return -1
    if total_batches <= warmup_batches: warmup_batches = 0

    with torch.no_grad():
        data_iter = iter(data_loader)
        batch_count, skipped_batches, processed_batches = 0, 0, 0
        pbar = tqdm(total=total_batches, desc=f"Measuring Inference Time ({model_type})")
        while processed_batches < total_batches:
            try: images, _ = next(data_iter); batch_count += 1; processed_batches += 1; pbar.update(1)
            except StopIteration: break
            except Exception as e: print(f"Err loading batch {processed_batches}: {e}. Skip."); skipped_batches += 1; pbar.update(1); continue
            try:
                images = list(img.to(pytorch_device) for img in images)
                batch_size = len(images);
                if batch_size == 0: continue
            except Exception as e: print(f"Err moving batch {processed_batches} to {pytorch_device}: {e}. Skip."); skipped_batches += 1; continue
            if batch_count <= warmup_batches:
                 try: _ = model(images)
                 except Exception as e: print(f"Err warmup batch {batch_count}: {e}")
                 continue
            try:
                start_time = time.time()
                _ = model(images)
                if pytorch_device.type == 'cuda': torch.cuda.synchronize()
                end_time = time.time()
                times.append(end_time - start_time); num_samples += batch_size
            except Exception as e: print(f"Err inference batch {batch_count}: {e}")
        pbar.close()
    if skipped_batches > 0: print(f"Warn: Skipped {skipped_batches} batches during inference.")
    batch_size_used = data_loader.batch_size if hasattr(data_loader, 'batch_size') and data_loader.batch_size else 1
    avg_batch_time = np.mean(times) if times else 0
    num_measured_batches = len(times)
    avg_image_time = (avg_batch_time / batch_size_used) * 1000 if batch_size_used > 0 and times else 0
    print(f"{model_type} - Avg Batch Time: {avg_batch_time:.4f}s, Avg Image Time: {avg_image_time:.2f}ms (over {num_measured_batches} batches)")
    return avg_image_time
# <<< END of unchanged measure_inference >>>

# --- Measure Inference Times ---
rcnn_inference_time_ms = -1
if 'rcnn_model_eval' in locals() and 'data_loader_test_rcnn' in locals() and data_loader_test_rcnn is not None and len(data_loader_test_rcnn) > 0:
     rcnn_inference_time_ms = measure_inference(rcnn_model_eval, data_loader_test_rcnn, DEVICE, 'Faster R-CNN')
elif 'rcnn_model_eval' not in locals(): print("Faster R-CNN model not evaluated, cannot measure time.")
else: print("Faster R-CNN test loader empty/unavailable, cannot measure time.")

yolo_inference_time_ms = -1
if 'yolo_model_eval' in locals() and 'data_loader_test_rcnn' in locals() and data_loader_test_rcnn is not None:
    # Note: Measuring YOLO time accurately might need a dedicated loop calling predict
    # This is an approximation using the val loop structure - Check Ultralytics docs for best practice
    print("Measuring YOLOv8 inference time (approximation using val structure)...")
    # Create a temporary dataloader using YOLO's test image dir if needed
    # Or adapt measure_inference for YOLO's predict method (more complex)
    # For now, just report N/A
    yolo_inference_time_ms = -1 # Mark as not accurately measured here
    print("YOLOv8 inference time measurement not fully implemented in this structure. Report N/A or implement dedicated predict loop.")
else: print("YOLOv8 model/test loader not available for inference time measurement.")


# --- Final Metrics Display ---
# Retrieve metrics safely
yolo_map = -1.0
if 'yolo_metrics' in locals() and hasattr(yolo_metrics, 'box') and hasattr(yolo_metrics.box, 'map'):
    yolo_map = yolo_metrics.box.map

rcnn_map = -1.0
if 'test_metrics' in locals() and isinstance(test_metrics, dict) and 'map' in test_metrics:
    map_tensor = test_metrics.get('map', torch.tensor(-1.0))
    if isinstance(map_tensor, torch.Tensor): rcnn_map = map_tensor.item()
    elif isinstance(map_tensor, (int, float)): rcnn_map = float(map_tensor)

print("\n--- Final Metrics (Approximate) ---")
print(f"YOLOv8 mAP@50-95: {yolo_map:.4f}" if yolo_map != -1.0 else "YOLOv8 mAP@50-95: N/A")
print(f"Faster R-CNN mAP@50-95: {rcnn_map:.4f}" if rcnn_map != -1.0 else "Faster R-CNN mAP@50-95: N/A")

print(f"\nYOLOv8 Inference Time: {yolo_inference_time_ms:.2f} ms" if yolo_inference_time_ms != -1 else "YOLOv8 Inference Time: Not Measured")
print(f"Faster R-CNN Inference Time: {rcnn_inference_time_ms:.2f} ms" if rcnn_inference_time_ms != -1 else "Faster R-CNN Inference Time: Not Measured")
if effdet_inference_time != -1.0:
     device_used_msg = "(CPU forced)" if physical_devices_gpu else "(CPU)"
     print(f"EfficientDet (TF Hub) Inference Time {device_used_msg}: {effdet_inference_time * 1000:.2f} ms")
else: print(f"EfficientDet (TF Hub) Inference Time: Not Measured or Failed")

# --- Download Instructions ---
print("\n--- Model Download Instructions ---")
print(f"After this notebook finishes and you 'Save Version' > 'Save & Run All (Commit)':")
print(f"1. Go to the completed notebook version viewer.")
print(f"2. Navigate to the 'Data' tab, then the 'Output' section.")
print(f"3. Look inside the '{MODEL_CHECKPOINTS_DIR.name}' folder.")
print(f"4. You should find the following model files (if training was successful):")

# Check final paths and report
if 'final_best_yolo_model_path' in locals() and final_best_yolo_model_path.exists():
    print(f"   - YOLOv8 Model: {final_best_yolo_model_path.name} (Path: {final_best_yolo_model_path})")
elif 'final_last_yolo_model_path' in locals() and final_last_yolo_model_path.exists():
     print(f"   - YOLOv8 Model: {final_last_yolo_model_path.name} (Last epoch) (Path: {final_last_yolo_model_path})")
else:
    print(f"   - YOLOv8 Model: Not found or not saved.")

if 'best_rcnn_model_path' in locals() and best_rcnn_model_path.exists():
    print(f"   - Faster R-CNN Model (State Dict): {best_rcnn_model_path.name} (Path: {best_rcnn_model_path})")
else:
    print(f"   - Faster R-CNN Model: Not found or not saved.")

print(f"5. For EfficientDet, no model file is saved by this script. Use the TF Hub URL: {hub_url if 'hub_url' in locals() else 'N/A'}")
print(f"6. Download the desired model file(s) using the download button next to the file name.")


print("\nScript finished.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pathlib import Path
import os # Make sure os is imported if not already
import torch # Ensure torch is imported for isinstance checks

# ==============================================================================
# 6. Results Visualization & Saving Plots
# ==============================================================================
print("\n--- Generating and Saving Evaluation Graphs ---")

# <<< Define a directory to save the plots >>>
PLOT_SAVE_DIR = Path('/kaggle/working/evaluation_plots')
PLOT_SAVE_DIR.mkdir(exist_ok=True) # Create directory if it doesn't exist
print(f"Plots will be saved to: {PLOT_SAVE_DIR}")

# --- Helper Function for Safe Data Retrieval ---
def get_metric(data, key, default=None):
    """Safely retrieve a metric, handling missing keys or non-scalar tensors."""
    if data is None: return default
    # Handle dictionary-like access first
    value = default
    if hasattr(data, 'get'):
        value = data.get(key, default)
    elif isinstance(data, dict):
         value = data.get(key, default)
    # Add specific handling for other types if needed (e.g., custom objects)

    if isinstance(value, torch.Tensor) and value.numel() == 1:
        return value.item()
    elif isinstance(value, (int, float)):
        return value
    return value # Return the original value if not scalar tensor/numeric

# --- Plot 1: Training Loss vs. Epoch ---
print("Plotting Training Loss vs. Epoch...")
fig_loss, ax_loss = plt.subplots(figsize=(12, 6)) # Get figure and axes objects
epochs_rcnn = range(1, len(rcnn_history.get('train_loss', [])) + 1) if 'rcnn_history' in locals() else []

# Faster R-CNN Loss
if epochs_rcnn and rcnn_history.get('train_loss'):
    ax_loss.plot(epochs_rcnn, rcnn_history['train_loss'], marker='o', linestyle='-', label='Faster R-CNN Train Loss')
else:
    print("  Faster R-CNN training history not found or empty.")

# YOLOv8 Loss (Requires reading results.csv)
yolo_results_path = None
if 'yolo_run_dir' in locals() and yolo_run_dir is not None and yolo_run_dir.exists():
    yolo_results_path = yolo_run_dir / 'results.csv'

if yolo_results_path and yolo_results_path.exists():
    try:
        # Check if yolo_df exists, otherwise load it
        if 'yolo_df' not in locals() or yolo_df is None:
             yolo_df = pd.read_csv(yolo_results_path)
             yolo_df.columns = yolo_df.columns.str.strip() # Strip whitespace here

        # Re-strip just in case it was loaded before but not stripped
        yolo_df.columns = yolo_df.columns.str.strip()

        if 'epoch' in yolo_df.columns and 'train/loss' in yolo_df.columns:
             yolo_epochs_plot = yolo_df['epoch'] + 1
             ax_loss.plot(yolo_epochs_plot, yolo_df['train/loss'], marker='s', linestyle='--', label='YOLOv8 Train Loss')
             if 'val/loss' in yolo_df.columns:
                  ax_loss.plot(yolo_epochs_plot, yolo_df['val/loss'], marker='x', linestyle=':', label='YOLOv8 Validation Loss')
        else:
            print(f"  'epoch' or 'train/loss' columns not found in {yolo_results_path}")
            print(f"  Available columns: {list(yolo_df.columns)}")

    except Exception as e:
        print(f"  Error reading or plotting YOLOv8 results.csv: {e}")
else:
    print("  YOLOv8 results.csv not found.")

ax_loss.set_title('Training Loss per Epoch')
ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.grid(True, which='both', linestyle='--', linewidth=0.5)
ax_loss.legend()
plt.tight_layout()

# <<< Save the loss plot >>>
loss_plot_path = PLOT_SAVE_DIR / 'training_loss_vs_epoch.png'
try:
    fig_loss.savefig(loss_plot_path, bbox_inches='tight', dpi=150)
    print(f"  Saved loss plot to: {loss_plot_path}")
except Exception as e:
    print(f"  Error saving loss plot: {e}")
plt.show()


# --- Plot 2: Validation mAP vs. Epoch ---
print("\nPlotting Validation mAP vs. Epoch...")
fig_map, ax_map = plt.subplots(figsize=(12, 6)) # Get figure and axes objects
epochs_rcnn_val = range(1, len(rcnn_history.get('val_map', [])) + 1) if 'rcnn_history' in locals() else []

# Faster R-CNN Validation mAP
if epochs_rcnn_val and rcnn_history.get('val_map') and any(m != -1.0 for m in rcnn_history['val_map']):
    ax_map.plot(epochs_rcnn_val, rcnn_history['val_map'], marker='o', linestyle='-', label='Faster R-CNN Val mAP@50-95')
    if rcnn_history.get('val_map_50') and any(m != -1.0 for m in rcnn_history['val_map_50']):
         ax_map.plot(epochs_rcnn_val, rcnn_history['val_map_50'], marker='o', linestyle='--', label='Faster R-CNN Val mAP@50')
else:
    print("  Faster R-CNN validation history not found, empty, or contains only default values.")

# YOLOv8 Validation mAP
yolo_map_col = 'metrics/mAP50-95(B)'
yolo_map50_col = 'metrics/mAP50(B)'

if yolo_results_path and yolo_results_path.exists():
    try:
        # Ensure yolo_df is loaded and columns stripped
        if 'yolo_df' not in locals() or yolo_df is None:
             yolo_df = pd.read_csv(yolo_results_path)
        yolo_df.columns = yolo_df.columns.str.strip()

        if 'epoch' in yolo_df.columns and yolo_map_col in yolo_df.columns:
             yolo_epochs_plot = yolo_df['epoch'] + 1
             ax_map.plot(yolo_epochs_plot, yolo_df[yolo_map_col], marker='s', linestyle='-', label='YOLOv8 Val mAP@50-95')
             if yolo_map50_col in yolo_df.columns:
                   ax_map.plot(yolo_epochs_plot, yolo_df[yolo_map50_col], marker='s', linestyle='--', label='YOLOv8 Val mAP@50')
             else:
                  print(f"   '{yolo_map50_col}' column not found for YOLOv8 mAP@50 plot.")
        else:
            print(f"  Required mAP columns ('{yolo_map_col}') not found in {yolo_results_path}")
            print(f"  Available columns: {list(yolo_df.columns)}")

    except Exception as e:
        print(f"  Error reading or plotting YOLOv8 results.csv for mAP: {e}")
# else: # Already printed message if file not found

ax_map.set_title('Validation mAP per Epoch')
ax_map.set_xlabel('Epoch')
ax_map.set_ylabel('mAP Score')
ax_map.set_ylim(bottom=0)
ax_map.grid(True, which='both', linestyle='--', linewidth=0.5)
ax_map.legend()
plt.tight_layout()

# <<< Save the mAP plot >>>
map_plot_path = PLOT_SAVE_DIR / 'validation_map_vs_epoch.png'
try:
    fig_map.savefig(map_plot_path, bbox_inches='tight', dpi=150)
    print(f"  Saved mAP plot to: {map_plot_path}")
except Exception as e:
    print(f"  Error saving mAP plot: {e}")
plt.show()


# --- Plot 3: Final Test Metrics Comparison (mAP & Inference Time) ---
print("\nPlotting Final Test Metrics Comparison...")
model_names = ['YOLOv8', 'Faster R-CNN', 'EfficientDet (TF Hub)']
map_scores = []
map_50_scores = []
inference_times_ms = []

# Get YOLOv8 metrics
yolo_final_map = -1.0
yolo_final_map50 = -1.0
if 'yolo_metrics' in locals() and yolo_metrics is not None and hasattr(yolo_metrics, 'box'):
    try:
        yolo_final_map = yolo_metrics.box.map
        yolo_final_map50 = yolo_metrics.box.map50
    except Exception as e:
        print(f"Could not extract YOLOv8 metrics: {e}")
map_scores.append(yolo_final_map if yolo_final_map != -1.0 else 0)
map_50_scores.append(yolo_final_map50 if yolo_final_map50 != -1.0 else 0)
yolo_inf_time = yolo_inference_time_ms if 'yolo_inference_time_ms' in locals() and yolo_inference_time_ms != -1 else 0
inference_times_ms.append(yolo_inf_time)

# Get Faster R-CNN metrics
rcnn_final_map = get_metric(locals().get('test_metrics'), 'map', default=-1.0)
rcnn_final_map50 = get_metric(locals().get('test_metrics'), 'map_50', default=-1.0)
map_scores.append(rcnn_final_map if rcnn_final_map != -1.0 else 0)
map_50_scores.append(rcnn_final_map50 if rcnn_final_map50 != -1.0 else 0)
rcnn_inf_time = rcnn_inference_time_ms if 'rcnn_inference_time_ms' in locals() and rcnn_inference_time_ms != -1 else 0
inference_times_ms.append(rcnn_inf_time)

# Get EfficientDet metrics
map_scores.append(0)
map_50_scores.append(0)
effdet_inf_time_sec = effdet_inference_time if 'effdet_inference_time' in locals() and effdet_inference_time != -1.0 else 0
inference_times_ms.append(effdet_inf_time_sec * 1000)

# Create Bar Chart
fig_metrics, axs_metrics = plt.subplots(1, 2, figsize=(15, 6)) # Get figure and axes objects

# mAP plot
x_pos = np.arange(len(model_names))
width = 0.35
rects1 = axs_metrics[0].bar(x_pos - width/2, map_scores, width, label='mAP@50-95')
rects2 = axs_metrics[0].bar(x_pos + width/2, map_50_scores, width, label='mAP@50')
axs_metrics[0].set_ylabel('mAP Score')
axs_metrics[0].set_title('Test Set mAP Comparison')
axs_metrics[0].set_xticks(x_pos)
axs_metrics[0].set_xticklabels(model_names, rotation=15, ha='right')
axs_metrics[0].legend()
axs_metrics[0].grid(True, axis='y', linestyle='--', linewidth=0.5)
axs_metrics[0].bar_label(rects1, padding=3, fmt='%.3f')
axs_metrics[0].bar_label(rects2, padding=3, fmt='%.3f')
axs_metrics[0].set_ylim(bottom=0)
axs_metrics[0].text(2, max(map_scores + map_50_scores) * 0.1 if max(map_scores + map_50_scores) > 0 else 0.01, '(N/A)*', ha='center', va='bottom', color='red', fontsize=10)
axs_metrics[0].text(0.5, -0.30, '*EfficientDet not evaluated on this dataset', ha='center', va='top', transform=axs_metrics[0].transAxes, fontsize=9) # Adjusted y position slightly

# Inference Time plot
axs_metrics[1].bar(model_names, inference_times_ms, color=['blue', 'orange', 'green'])
axs_metrics[1].set_ylabel('Average Inference Time (ms) per Image')
axs_metrics[1].set_title('Inference Speed Comparison')
axs_metrics[1].set_xticks(x_pos) # Ensure x-ticks match positions
axs_metrics[1].set_xticklabels(model_names, rotation=15, ha='right')
axs_metrics[1].grid(True, axis='y', linestyle='--', linewidth=0.5)
max_inf_time = max(inference_times_ms) if inference_times_ms else 1 # Avoid division by zero
for i, v in enumerate(inference_times_ms):
    axs_metrics[1].text(i, v + max_inf_time * 0.02, f"{v:.1f}", ha='center', va='bottom')

plt.suptitle('Model Performance Comparison', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# <<< Save the comparison plot >>>
comparison_plot_path = PLOT_SAVE_DIR / 'final_metrics_comparison.png'
try:
    fig_metrics.savefig(comparison_plot_path, bbox_inches='tight', dpi=150)
    print(f"  Saved comparison plot to: {comparison_plot_path}")
except Exception as e:
    print(f"  Error saving comparison plot: {e}")
plt.show()


# --- Plot 4: Confusion Matrices ---
print("\nPlotting Confusion Matrices (where available)...")

# YOLOv8 Confusion Matrix
yolo_cm = None
if 'yolo_metrics' in locals() and yolo_metrics is not None and hasattr(yolo_metrics, 'confusion_matrix'):
    yolo_cm = yolo_metrics.confusion_matrix

if yolo_cm is not None:
    print("  Plotting YOLOv8 Confusion Matrix...")
    cm_png_path = None
    if 'yolo_run_dir' in locals() and yolo_run_dir is not None:
         cm_png_path = yolo_run_dir / 'confusion_matrix.png' # Default save name

    try:
        # Plotting using ultralytics method also saves it
        yolo_cm.plot(normalize=False, save_dir=str(yolo_run_dir) if yolo_run_dir else '.', names=CLASSES)
        print(f"    Ultralytics confusion matrix plot saved in: {yolo_run_dir}")

        # Try to display the saved plot
        if cm_png_path and cm_png_path.exists():
             img = plt.imread(cm_png_path)
             fig_cm_yolo_ultralytics, ax_cm_yolo_ultralytics = plt.subplots(figsize=(10, 8))
             ax_cm_yolo_ultralytics.imshow(img)
             ax_cm_yolo_ultralytics.axis('off')
             ax_cm_yolo_ultralytics.set_title("YOLOv8 Confusion Matrix (from ultralytics plot)")

             # <<< Save the displayed ultralytics plot figure >>>
             yolo_cm_display_path = PLOT_SAVE_DIR / 'yolov8_confusion_matrix_ultralytics.png'
             try:
                 fig_cm_yolo_ultralytics.savefig(yolo_cm_display_path, bbox_inches='tight', dpi=150)
                 print(f"    Saved display copy of Ultralytics CM plot to: {yolo_cm_display_path}")
             except Exception as e:
                 print(f"    Error saving display copy of Ultralytics CM plot: {e}")
             plt.show()
        else:
            print("    Could not find saved YOLOv8 confusion matrix plot from Ultralytics. Plotting raw matrix.")
            # Fallback: Plot the raw matrix using seaborn
            fig_cm_yolo_raw, ax_cm_yolo_raw = plt.subplots(figsize=(10, 8))
            cm_array = yolo_cm.matrix
            cm_labels = CLASSES + ['Background FP'] if cm_array.shape[0] == len(CLASSES) + 1 else CLASSES
            sns.heatmap(cm_array, annot=True, fmt='d', cmap='Blues',
                        xticklabels=cm_labels, yticklabels=cm_labels, ax=ax_cm_yolo_raw)
            ax_cm_yolo_raw.set_title('YOLOv8 Confusion Matrix (Raw)')
            ax_cm_yolo_raw.set_xlabel('Predicted Class')
            ax_cm_yolo_raw.set_ylabel('True Class')
            plt.tight_layout()

            # <<< Save the raw seaborn CM plot >>>
            yolo_cm_raw_path = PLOT_SAVE_DIR / 'yolov8_confusion_matrix_raw.png'
            try:
                 fig_cm_yolo_raw.savefig(yolo_cm_raw_path, bbox_inches='tight', dpi=150)
                 print(f"    Saved raw CM plot to: {yolo_cm_raw_path}")
            except Exception as e:
                 print(f"    Error saving raw CM plot: {e}")
            plt.show()

    except Exception as e:
        print(f"    Error plotting YOLOv8 confusion matrix: {e}")
else:
    print("  YOLOv8 confusion matrix data not available.")

# Faster R-CNN Confusion Matrix
print("\n  Faster R-CNN: Standard confusion matrix not generated by default evaluation.")
print("    Requires custom matching of predictions to ground truth based on IoU.")

# EfficientDet Confusion Matrix
print("\n  EfficientDet (TF Hub): Confusion matrix not applicable (trained on different dataset/classes).")


print("\n--- Visualization Complete ---")
print(f"All generated plots saved in: {PLOT_SAVE_DIR}")
```
